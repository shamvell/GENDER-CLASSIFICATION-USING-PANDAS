```python
#to check yhe ouliers
columns=["long_hair","forehead_width_cm","forehead_height_cm","nose_wide","nose_long","lips_thin","distance_nose_to_lip_long"]
for col in columns :
    Q1=df[col].quantile(0.25)
    Q2=df[col].quantile(0.75)
    IQR=Q2-Q1
    LOWER_BOUND=Q1-1.5*IQR
    UPPER_BOUND=Q2+1.5*IQR
    outliers=df[(df[col]<LOWER_BOUND )|(df[col]>UPPER_BOUND)]
    print(outliers.shape[0])
```

```python
#clip to remove the outliers
for col in columns:
    Q1=df[col].quantile(0.25)
    Q3=df[col].quantile(0.75)
    IQR=Q3-Q1
    LOWER_BOUND=Q1-1.5*IQR
    UPPER_BOUND=Q3+1.5*IQR
    outliers=df[(df[col]<LOWER_BOUND) | (df[col]>UPPER_BOUND)]
    df[col]=df[col].clip(lower=LOWER_BOUND,upper=UPPER_BOUND)
```

```python
columns=["long_hair","forehead_width_cm","forehead_height_cm","nose_wide","nose_long","lips_thin","distance_nose_to_lip_long"]
for col in columns :
    Q1=df[col].quantile(0.25)
    Q2=df[col].quantile(0.75)
    IQR=Q2-Q1
    LOWER_BOUND=Q1-1.5*IQR
    UPPER_BOUND=Q2+1.5*IQR
    outliers=df[(df[col]<LOWER_BOUND )|(df[col]>UPPER_BOUND)]
    print(outliers.shape[0])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#vertical bar plot


avg_by_gender = df.groupby("gender")[["long_hair", "forehead_width_cm", "forehead_height_cm",
                                      "nose_wide", "nose_long", "lips_thin", 
                                      "distance_nose_to_lip_long"]].median().reset_index()
colurs="red","blue","pink","violet","orange"
# Plot
plt.figure(figsize=(12, 7))
avg_by_gender.set_index("gender").plot(kind='bar')
plt.title("Average Facial Features by Gender")
plt.ylabel("Average Value")
plt.xlabel("Gender")

plt.tight_layout()
plt.show()
```

```python
 #multiple boxplot showing the distribution
plt.figure(figsize=(15,5))
sns.boxplot(data=df)
plt.title("outliers of my dataset")
plt.show()
```

```python
#scattered plot
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='forehead_height_cm', y='forehead_width_cm', hue='gender')
plt.title("long hair vs forehead_width_cm")
plt.show()
```

```python
# Use a count plot

df_sorted = df.sort_values(by='gender')

plt.figure(figsize=(3,4))
sns.lineplot(x=range(len(df_sorted)), y=df_sorted['gender'].astype('category').cat.codes,
             marker='x', linestyle='-')

plt.title("Sorted Gender Trend")
plt.xlabel("Index (Sorted by Gender)")
plt.ylabel("Gender (coded)")
plt.yticks([0, 1], df_sorted['gender'].astype('category').cat.categories)
plt.show()
```

```python
# Pie Chart of Gender Count
gender_counts = df["gender"].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Gender Distribution")
plt.tight_layout()
plt.show()
```

```python
#Histogram of gender specification
plt.figure(figsize=(10, 5))
plt.hist(df["gender"], bins=10, edgecolor="black")
plt.title("average of gender specfication")
plt.xlabel("value")
plt.ylabel("Number of gender")
plt.tight_layout()
plt.show()
```

```python
#long_hair feature varies between genders.
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(2, 6))
sns.boxplot(x="gender", y="long_hair", data=df)
plt.title("Distribution of Long Hair Indicator by Gender")
plt.ylabel("value")
plt.tight_layout()
plt.show()
```

```python
#box plot
#forehead_width_cm values vary for each gender category.
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(2, 6))
sns.boxplot(x="gender", y="forehead_width_cm", data=df)
plt.title("forehead_width_cm")
plt.ylabel("value")
plt.tight_layout()
plt.show() 
```

```python
#If you actually want counts in order
plt.figure(figsize=(8, 5))
sns.countplot(x='gender', data=df, order=sorted(df['gender'].unique()))
plt.title("Gender Distribution (Sorted)")
plt.show()
```

```python
#Violin Plot – Shows distribution shape + median
plt.figure(figsize=(2, 6))
sns.violinplot(x="gender", y="forehead_width_cm", data=df)
plt.title("Forehead Width by Gender")
plt.ylabel("Value (cm)")
plt.tight_layout()
plt.show()
```

```python
#Strip Plot – Shows all data points
plt.figure(figsize=(2, 6))
sns.stripplot(x="gender", y="forehead_width_cm", data=df, jitter=True)
plt.title("Forehead Width by Gender")
plt.ylabel("Value (cm)")
plt.tight_layout()
plt.show()
```

```python
#Bar Plot (Mean/Median) – Shows average instead of spread

plt.figure(figsize=(4, 6))
sns.barplot(x="gender", y="forehead_width_cm", data=df, estimator="median")
plt.title("Median Forehead Width by Gender")
plt.ylabel("Value (cm)")
plt.tight_layout()
plt.show()
```
